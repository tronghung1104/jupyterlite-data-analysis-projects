# Keyword Dashboard for Content Optimisation

## Introduction

The project is a keyword report sample that was ordered by the company. It is served as an analytical dashboard where the marketing leader can access the data of each keyword and makes decision with which keywords he/she can optimise for the app listing page of the company app.

The data of each keyword including:
- `Keyword`: the word or phrase that users searched for the app on the app store
- `Pageviews`: the number of pageviews generated by the keyowrd
- `Add App click`: the number of clicks of CTA button that leads to an app install
- `Average CTR`: The click through rate of the keyword
- `Average Position`: the average position of the app by the keyword

## Preparation

The whole process of developing this dashboard including the developement of the following sheets:
- Get the dataset we need for the dashboard (`Applisting keyword-data` sheet)
- Build the query helper to retrieve data based on the filter of the dashboard (`_query_helper` sheet)
- Create a data validation as inputs for the keyword filter and order by mertric (`_data_validation` sheet)
- Finally, we build the dashboard inlcuding: filter, table and cards for data visuallisation (`BOGO-Applisting keyword-data_cleaned` sheet)
